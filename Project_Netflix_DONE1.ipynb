### Read the CSV Files (netflix_data.csv)

`Use professional method (using os library)`

```python
## Get the path and read the CSV Files using pandas --> DataFrame name is (df_netflix) --> show the head of DF
path=os.path.join(os.getcwd(),'netflix_data.csv')
df=pd.read_csv(path)
df
```

### `Explore the (type) column of the above DF, get the unique values, and create CountPlot `

```python
## Get the unqiue values of the (type) column of (df_netflix)
df_w=df['type'].unique
df_w
```

```python
## Create a CountPlot for this columns
## Don't forget the title and axis
  
plt.figure(figsize=(10, 6))   

sns.countplot(x='type', data=df)
plt.title('count for each category in type colun', fontsize=14)
plt.xlabel('type', fontsize=10)
plt.ylabel('Count', fontsize=10)

plt.show()
```

### `Slice the (df_netflix) DF, by getting only type=='Movie'`

* `I want you to get a DF, sliced from main DF (df_netflix), by getting it when (type='Movie')`
* `Assign the new DF named (df_movies) `

```python
## slice the (df_netflix), when (type=='Movie'),  getting a new DF named (df_movied)
## Show the head of the new DF
## Aftrer doing that, check the unique values again of the column (type), it must contains (Movie) 
df_new=df[df['type'] == 'Movie']
df_new
```

```python
df_new['type'].unique()
```

### `Explore the new DF, it contains two columns (release_year, durations)`
`Create a scatter plot of these two columns, from (df_movies) DataFrame`

```python
## Create a scatetr plot of (release_year, durations) columns -- use any method and any library
## Don't forget the title and axis labels --> make your graphs interesting and readable
df_m=df[['release_year','duration']]
df_m
df_m.plot.scatter(x='release_year',y='duration',c='red',grid=5)  
plt.title('relation between durations and release_years')
```

### `Going Deeper, what about getting short movies`

`Slice from (df_movies) the short movies which duration is less than 60 minutes --> new DF (df_short)`

```python
## slice from (df_movies) the short movies which their duration < 60 minutes
## Assign the new DF to variable named (df_short)
## show the head of new DF (df_short)

df_s=df_new[df_new['duration']<60]
df_s.head()
```

### `From (df_short), get only the below columns (slice the desired columns only)`

* `Get these columns only : ['title', 'country', 'genre', 'release_year', 'duration']`
* `Assign the DF name to the same name (df_short), I mean override or (inplace=True)`

```python
## getting the desired columns only ['title', 'country', 'genre', 'release_year', 'duration'] from (df_short)
## Show the Head  again after selecting the desired columns only 
df_s=df_s[['title','country','genre','release_year','duration']]
```

### `Create Scatter plot but make (hue='genre'), it is prefreable here to use (seaborn)`
* `Scatter plot between (releaes_year) & (duration) columns from (df_short)`
* `you can use the relplot in seaborn`

```python
## Scatter plot between (releaes_year) & (duration) columns from (df_short)
## Don't forget the axis labels and title 

plt.figure(figsize=(10, 6))   

sns.scatterplot(x='release_year', y='duration', data=df_s,hue='genre',alpha=1)

## title and axis
plt.title('Relation between release years and durations for short movies', fontsize=14)
plt.xlabel('Relation Years', fontsize=14)
plt.ylabel('Durations', fontsize=14)
plt.grid(5)
```
